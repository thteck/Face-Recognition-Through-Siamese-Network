```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
epochs = np.arange(1, 11)  # Assuming you have 10 epochs

# Training loss data for each model
resnet50_loss = [0.8282, 0.4027, 0.2886, 0.224, 0.1804, 0.1785, 0.1337, 0.1301, 0.1238, 0.113]
vgg16_loss = [11.6574, 4.9036, 2.5705, 2.3846, 1.6938, 1.4355, 1.0411, 0.8225, 0.6725, 0.6114]
inceptionv3_loss = [0.8559, 0.4236, 0.3568, 0.293, 0.3049, 0.2607, 0.2549, 0.221, 0.1972, None]

# Validation loss data for each model
resnet50_val_loss = [0.7351, 0.434, 0.3421, 0.3895, 0.321, 0.3635, 0.3028, 0.3295, 0.2856, 0.2635]
vgg16_val_loss = [7.5087, 4.1229, 4.0012, 3.1819, 2.4545, 2.6313, 1.4873, 1.9587, 1.5528, 0.8685]
inceptionv3_val_loss = [0.6153, 0.4654, 0.3649, 0.2773, 0.3436, 0.2908, 0.3328, 0.3656, 0.3027, None]
```

```python
# Plot training loss
plt.figure(figsize=(12, 6))
plt.plot(epochs, resnet50_loss, marker='o', label='B', linestyle='-', color='blue')
plt.plot(epochs, vgg16_loss, marker='o', label='C', linestyle='-', color='green')
plt.plot(epochs, inceptionv3_loss, marker='o', label='D', linestyle='-', color='red')

# Plot validation loss
plt.plot(epochs, resnet50_val_loss, marker='o', linestyle='--', color='blue')
plt.plot(epochs, vgg16_val_loss, marker='o', linestyle='--', color='green')
plt.plot(epochs, inceptionv3_val_loss, marker='o', linestyle='--', color='red')

# Set plot labels and title
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
